```python
#引入相关包
import scipy.io as scio
import numpy as np
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report
from sklearn.decomposition import  PCA
import pandas as pd
```

```python
data = scio.loadmat('Wine359_2203.mat')
arrMat = data['X']
label = data['Y']
```

```python
#查看数据共有2203维特征
print("array:{},label:{}".format(np.shape(arrMat),np.shape(label)))
```

```python
#通过查看标签，可知要实现多分类任务
label
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(arrMat,label,test_size=0.2,random_state=0)
#x_train,x_test=train_test_split(arrMat,test_size=0.2,random_state=0)
x_train.shape,x_test.shape
#print(x_train)
#print(x_test)
```

```python
# 2 对数据进行降维，维度降低到20维度
estimator = PCA(n_components=15)
pca_x_train = estimator.fit_transform(x_train)
pca_x_test = estimator.transform(x_test)
```

```python
# 3 使用默认配置的支持向量机进行学习和预测未降维的数据
svc = LinearSVC()
```

```python
# 学习
svc.fit(x_train, y_train)
```

```python
# 预测
y_predict = svc.predict(x_test)
```

```python
# 3.2 使用默认配置的支持向量机学习和预测降维后的数据
pca_svc = LinearSVC()
```

```python
# 学习
pca_svc.fit(pca_x_train, y_train)
pca_y_predict = pca_svc.predict(pca_x_test)
```

```python
# 4 模型评估
print("原始数据的准确率：", svc.score(x_test, y_test))
print("其他评分：\n", classification_report(y_test, y_predict, target_names=np.arange(4).astype(str)))
print("降维后的数据准确率:", pca_svc.score(pca_x_test, y_test))
print("其他评分：\n", classification_report(y_test, pca_y_predict, target_names=np.arange(4).astype(str)))
```
